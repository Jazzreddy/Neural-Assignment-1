```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.losses import MeanSquaredError, CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.callbacks import TensorBoard
import datetime
```

```python
# Task 1: Tensor Manipulations & Reshaping
print("Task 1: Tensor Manipulations & Reshaping")

# Creating a random tensor of shape (4,6)
tensor = tf.random.uniform(shape=(4, 6))
print("Original Tensor:", tensor.numpy())

# Displaying rank and shape of the tensor
print("Rank:", tf.rank(tensor).numpy(), "Shape:", tensor.shape)

# Reshaping the tensor to a 3D shape (2,3,4)
reshaped_tensor = tf.reshape(tensor, (2, 3, 4))

# Transposing the tensor along specific dimensions
transposed_tensor = tf.transpose(reshaped_tensor, perm=[1, 0, 2])
print("Reshaped Shape:", reshaped_tensor.shape, "Transposed Shape:", transposed_tensor.shape)

# Broadcasting: Creating a small tensor and expanding it to match another tensor's shape
small_tensor = tf.random.uniform(shape=(1, 6))
broadcasted_tensor = tf.broadcast_to(small_tensor, [4, 6])

# Performing element-wise addition after broadcasting
sum_tensor = tensor + broadcasted_tensor
print("Broadcasted Tensor:", broadcasted_tensor.numpy())
print("Sum Tensor:", sum_tensor.numpy())
```

```python
# Task 2: Loss Functions & Hyperparameter Tuning
print("\nTask 2: Loss Functions & Hyperparameter Tuning")

# Defining true labels and predicted values from two different models
y_true = np.array([1, 0, 0, 1])
y_pred1 = np.array([0.9, 0.1, 0.2, 0.8])
y_pred2 = np.array([0.8, 0.2, 0.3, 0.7])

# Initializing Mean Squared Error and Categorical Crossentropy loss functions
mse = MeanSquaredError()
cce = CategoricalCrossentropy()

# Calculating loss values for both predictions using MSE and CCE
mse_loss1 = mse(y_true, y_pred1).numpy()
mse_loss2 = mse(y_true, y_pred2).numpy()
cce_loss1 = cce(y_true, y_pred1).numpy()
cce_loss2 = cce(y_true, y_pred2).numpy()

# Printing the loss values to compare performance
print(f"MSE Loss 1: {mse_loss1}, MSE Loss 2: {mse_loss2}")
print(f"CCE Loss 1: {cce_loss1}, CCE Loss 2: {cce_loss2}")

# Visualizing the loss function values using a bar plot
plt.bar(['MSE1', 'MSE2', 'CCE1', 'CCE2'], [mse_loss1, mse_loss2, cce_loss1, cce_loss2])
plt.title("Loss Function Comparison")
plt.show()
```

```python

# Task 3: Train a Model with Different Optimizers
def train_mnist(optimizer, epochs=5):
    """
    Trains a simple neural network on the MNIST dataset using the specified optimizer.

    Args:
        optimizer: Optimizer for training (Adam or SGD).
        epochs: Number of epochs to train the model.

    Returns:
        Training history containing accuracy and loss values.
    """
    # Loading the MNIST dataset and normalizing pixel values
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalization

    # Defining a simple feedforward neural network
    model = Sequential([
        Flatten(input_shape=(28, 28)),  # Flattening the 28x28 image into a 1D vector
        Dense(128, activation='relu'),  # Hidden layer with ReLU activation
        Dense(10, activation='softmax')  # Output layer with softmax activation
    ])

    # Compiling the model with the given optimizer and loss function
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    # Training the model and storing history for comparison
    history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epochs, verbose=2)
    return history

# Training and comparing models with different optimizers
print("\nTraining with Adam Optimizer...")
adam_history = train_mnist(Adam())

print("\nTraining with SGD Optimizer...")
sgd_history = train_mnist(SGD())

# Plotting accuracy over epochs to compare optimizers
plt.plot(adam_history.history['accuracy'], label='Adam')
plt.plot(sgd_history.history['accuracy'], label='SGD')
plt.title('Optimizer Comparison - Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python

# Task 4: Train a Neural Network and Log to TensorBoard

# Setting up the logging directory for TensorBoard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

print("\nTask 4: Training with TensorBoard Logging")

# Loading and normalizing the MNIST dataset again
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Defining the neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Flattening the image
    Dense(128, activation='relu'),  # Hidden layer
    Dense(10, activation='softmax')  # Output layer
])

# Compiling the model with Adam optimizer and sparse categorical crossentropy loss
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training the model with TensorBoard callback enabled
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, callbacks=[tensorboard_callback], verbose=2)

# Instruction to launch TensorBoard for visualizing training logs
print("Launch TensorBoard using: tensorboard --logdir=logs/fit")
```
